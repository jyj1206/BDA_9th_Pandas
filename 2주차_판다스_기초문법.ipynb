### 파생변수 만들기

```python
df_score = pd.DataFrame({
    '수업' : ['국어', '영어', '수학', '과학'],
    '수행평가' : [30, 15, 25, 5],
    '중간고사' : [30, 20, 15, 10],
    '기말고사' : [40, 25, 20, 10]
})
```

```python
df_score
```

```python
# 파생변수 생성
# 총점 = 수행평가 + 중간고사 + 기말고사
df_score['총점'] = df_score['수행평가'] + df_score['중간고사'] + df_score['기말고사']
```

```python
df_score
```

```python
# 넘파이 where를 이용한 파생변수 생성
df_score['이수여부'] = np.where(df_score['총점'] < 60, '재이수', '이수')
```

```python
df_score
```

```python
# 빈도수 카운팅
df_score['이수여부'].value_counts()
```

```python
dia['cut'].value_counts()
```

### 데이터 전처리 및 추출

#### query() - 데이터 추출

```python
# query(특정 요구조건)
# 문자열 데이터
dia.query('cut == "Premium"')
```

```python
# 수치형 데이터
dia.query('carat >= 1')
```

```python
# 다중 조건
dia.query('cut == "Premium" & carat >= 1')
```

```python
# 다중 조건
dia.query('depth < 50 | carat >= 1')
```

```python
# in 조건
dia.query('cut in ["Premium", "Ideal"]')
```

#### df[[불린참값]] - 데이터 추출

```python
dia[dia['clarity'] == 'VS2']
```

```python
# 데이터프레임 재구조화
dia[(dia['clarity'] == 'VS2') | (dia['clarity'] == 'SI1')]['cut'].value_counts()
```
