```python
##  컬럼명 변경
# 먼저 데이터 복사
mpg_copy1 = mpg.copy() # 변수 교체를 이용한 문법 적용
mpg_copy2 = mpg.copy() # inplace = True 문법 적용
```

```python
# 변수 교체를 이용해 컬럼명 변경 (mpg -> fuel_efficiency)
mpg_copy1 = mpg_copy1.rename(columns = {'mpg' : 'fuel_efficiency'})
```

```python
# 결과 확인
mpg_copy1
```

```python
# inplace = True 문법 적용
mpg_copy2.rename(columns = {'mpg' : 'fuel_efficiency'}, inplace = True)
```

```python
# 결과 확인
mpg_copy2
```

```python
## 파생변수 만들기
# 변수처럼 사용, 배기량 대비 출력 (horsepoer / displacement)
mpg['power_to_displacement'] = mpg['horsepower'] / mpg['displacement']
```

```python
# 결과 확인
mpg
```

```python
## 넘파이를 사용한 파생변수 (np.where)
# 배기량 대비 출력(power_to_displacement)이 평균보다 높으면 좋음, 그것보다 낮으면 안좋음
mpg['performance'] = np.where(mpg['power_to_displacement'] > mpg['power_to_displacement'].mean(), "좋음", "안좋음")
```

```python
# 결과 확인
mpg
```

```python
## 빈도수 출력
# performance의 빈도수 출력
mpg['performance'].value_counts()
```

```python
# 데이터 추출
# query를 이용한 특정 조건의 수치형 데이터 추출 (weight가 3500보다 작은 데이터)
mpg.query("weight < 3500")
```

```python
# query를 이용한 특정 열의 문자열 추출 (제조지가 미국인 데이터)
mpg.query("origin == 'usa'")
```

```python
# query를 이용한 다중 조건 데이터 추출 (cylinders 수가 4개이면서 performance가 좋음인 데이터)
mpg.query('cylinders == 4 & performance == "좋음"')
```

```python
# query를 이용한 특정 값들 중 하나를 가지는 데이터 추출 (제조지가 일본, 유럽 중 하나인 데이터)
mpg.query("origin in ['japan', 'europe']")
```

```python
# df[[불린참값]] 형태로 데이터 추출 (제조지가 미국인 데이터)
mpg[mpg['origin'] == 'usa']
```

```python
## 데이터 정렬하기
# 기준컬럼 1개의 열을 기준으로 오름차순 정렬 (acceleration)
mpg.sort_values('acceleration')
```

```python
# 기준컬럼 1개의 열을 기준으로 내림차순 정렬 (acceleration)
mpg.sort_values('acceleration', ascending=False)
```

```python
# 기준 컬럼 2개 이상의 열로 정렬하기 (acceleration 오름차순으로 정렬하고(낮을 수록 좋음), weight를 내림차순으로 정렬)
mpg.sort_values(["acceleration", "weight"], ascending = [True, False])
```

```python
## 데이터 집계
# Grouby를 이용하여 데이터 집계 (제조지를 기준으로 group)
mpg_gr = mpg.groupby('origin')
```

```python
# 결과 출력
for i, j in mpg_gr:
  print(i, j)
```

```python
# get_group을 이용해 특정 값의 그룹한 값들의 데이터프레임 반환 (제조지가 일본인 데이터)
mpg_gr.get_group('japan')
```

```python
# group by로 2개 이상의 열로 묶기 (제조지와 performance로 묶기)
mpg_gr2 = mpg.groupby(['origin', 'performance'])
```

```python
# get_group을 이용해 특정 값의 그룹한 값들의 데이터프레임 반환 (제조지가 일본이면서 performance가 나쁨인 데이터)
mpg_gr2.get_group(('japan', '안좋음'))
```

```python
# groupby 후 agg를 이용한 통계치 확인 (제조지마다 자동차의 weight의 평균과 차량 연식의 평균)
mpg_gr.agg({'weight' : 'mean', 'model_year' : 'mean'})
```

```python
# 특정 함수를 만들어 적용 agg 적용
# 함수 정의 (범위)
def min_max(x):
    return x.max()- x.min()
```

```python
# 함수 적용 (제조지별 weight와 acceleration의 범위)
mpg[['origin','weight', 'acceleration']].groupby('origin').agg(min_max)
```

```python
## 구간 나누기
# qcut을 이용한 분위수에 따른 구간 나누기 (weight에 따라 3구간으로 나누기)
res_qcut = pd.qcut(mpg['weight'], q = 3, labels = ['light', 'middle', 'heavy'])
```

```python
# 결과 출력
res_qcut
```

```python
# 구간화된 빈도수 확인
res_qcut.value_counts()
```

```python
# cut을 이용한 일정한 크기로 구간 나누기 (weight에 따라 3구간으로 나누기)
res_cut = pd.cut(mpg['weight'], bins = 3, labels = ['light', 'middle', 'heavy'])
```

```python
res_cut
```

```python
res_cut.value_counts()
```

```python
## 전체 데이터에 전처리 적용
# apply 함수 사용
# 함수 정의 (연비에 따라 등급 나누기)
def mpg_to_grade(mpg):
    if mpg >= 25:
        return 'A'
    elif mpg >= 20:
        return 'B'
    else:
        return 'C'
```

```python
# apply 적용
mpg['mpg_grade'] = mpg['mpg'].apply(mpg_to_grade)
```

```python
# 결과 출력 (빈도수로 확인)
mpg['mpg_grade'].value_counts()
```

```python
# assign 함수 사용 (기종별 모델의 현재를 기준으로 나이 확인)
mpg.assign(age = 2024 - (1900 +mpg['model_year']))
```

```python
## 데이터 병합
# 데이터 준비
df_A = pd.DataFrame({
    '프로그래밍 언어': ['Python', 'Java', 'C++', 'JavaScript', 'Ruby'],
    '개발 년도': [1991, 1995, 1983, 1995, 1995]
})

df_B = pd.DataFrame({
    '프로그래밍 언어': ['Python', 'Java', 'C++', 'JavaScript', 'Go'],
    '2023년 점유율': [29.9, 19.1, 7.5, 8.2, 4.6]
})
```

```python
# concat을 이용해 병합 (세로로 병합)
pd.concat([df_A, df_B])
```

```python
# merge를 이용해 프로그래밍 언어를 기준으로 inner 조인
pd.merge(df_A, df_B, how = 'inner', on = '프로그래밍 언어')
```
